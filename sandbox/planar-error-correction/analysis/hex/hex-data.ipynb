```python
# Estimate the total logical error as the sum of the Z and X
# basis results. This estimate is px + pz + 2*py. Assuming that
# py ~ O(px^2) we should be making a negligible error by approximating
# the total error in this way.
def total_error(zbasislists, xbasislists):
    physical = []
    logical = []
    xlookup = {u: v for u, v in zip(xbasislists[0], xbasislists[1])}
    for i in range(len(zbasislists[0])):
        p = zbasislists[0][i]
        p1z = zbasislists[1][i]
        if p in xlookup:
            physical.append(p)
            logical.append(p1z + xlookup[p])
    return physical, logical
```
